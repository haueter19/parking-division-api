```python
permit_access_range = pd.read_sql("""
    WITH

    -- ── 1. Normalize only the rows we care about ──────────────────────────────
    relevant_events AS (
        SELECT
            CONVERT(DATETIME, CONVERT(VARCHAR, CAST(dtDate AS DATE), 120) + ' ' + sTime) AS timestamp,
            lInstanceId,
            sInstanceDetails2 AS GroupName,
            sInstanceDetails1 AS CustomerName,
            CASE
                WHEN ePropertyType = 2   THEN 'Group assignment'
                WHEN ePropertyType = 104 THEN 'dtStartDate'
                WHEN ePropertyType = 110 THEN 'bManuallyBlockedOnEntries'
                ELSE CAST(ePropertyType AS VARCHAR(4))
            END AS ePropertyType,
            sPropertyOldValue AS old_value,
            sPropertyNewValue AS new_value,
            ROW_NUMBER() OVER (
                PARTITION BY lInstanceId
                ORDER BY CONVERT(DATETIME, CONVERT(VARCHAR, CAST(dtDate AS DATE), 120) + ' ' + sTime)
            ) AS rn
        FROM InstanceModificationLog
        WHERE
            ePropertyType IN (110, 2, 104)
            AND (sInstanceDetails1 NOT LIKE '%PU Employees%' AND sInstanceDetails1 NOT LIKE '%Brayton%')
            --AND lInstanceId = 16982
            --AND dtDate >= '2020-01-01'
            AND sInstanceDetails2 != '' -- must have a GroupName or we can't connect to access policy (only 1 like this)
    ),
    
    -- ── 1b. Resolve effective start date per card ─────────────────────────────
    card_start_dates AS (
        SELECT
            re.lInstanceId,
            COALESCE(
                CONVERT(DATE, REPLACE(MAX(CASE WHEN re.ePropertyType = 'dtStartDate' THEN re.new_value END), '-', ' '), 106),
                sc.dtStartDate
            ) AS effective_start_date,
            --CONCAT('L300', REPLACE(sc.sNumber, '-', '')) AS CardNumber
            sc.sNumber AS CardNumber
        FROM relevant_events re
        LEFT JOIN Ecounting.dbo.SPMCards sc ON sc.lId = re.lInstanceId
        GROUP BY re.lInstanceId, sc.dtStartDate, sc.sNumber
    ),
    
    -- ── 2. Tag each row with point-in-time signals ────────────────────────────
    signals AS (
        SELECT
            timestamp,
            lInstanceId,
            CustomerName,
            rn,
            CASE
                WHEN ePropertyType = 'bManuallyBlockedOnEntries'
                THEN CASE WHEN new_value = 'True' THEN 1 ELSE 0 END
            END AS blocked_signal,
            CASE
                WHEN ePropertyType = 'Group assignment' THEN GroupName
            END AS group_signal,
            COUNT(CASE WHEN ePropertyType = 'bManuallyBlockedOnEntries' THEN 1 END)
                OVER (PARTITION BY lInstanceId ORDER BY timestamp, rn
                      ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS blocked_grp,
            COUNT(CASE WHEN ePropertyType = 'Group assignment' THEN 1 END)
                OVER (PARTITION BY lInstanceId ORDER BY timestamp, rn
                      ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS group_grp
        FROM relevant_events
    ),
    
    -- ── 3. Carry last known values forward ────────────────────────────────────
    running_state AS (
        SELECT
            timestamp,
            lInstanceId,
            CustomerName,
            rn,
            MAX(blocked_signal) OVER (PARTITION BY lInstanceId, blocked_grp) AS is_blocked,
            MAX(group_signal)   OVER (PARTITION BY lInstanceId, group_grp)   AS current_group
        FROM signals
    ),
    
    -- ── 4. Derive access state at each event ──────────────────────────────────
    access_state AS (
        SELECT
            timestamp,
            lInstanceId,
            CustomerName,
            COALESCE(is_blocked, 0) AS is_blocked,
            COALESCE(current_group, (
                SELECT TOP 1 sInstanceDetails2
                FROM InstanceModificationLog sub
                WHERE sub.lInstanceId = running_state.lInstanceId
                ORDER BY CONVERT(DATETIME, CONVERT(VARCHAR, CAST(sub.dtDate AS DATE), 120) + ' ' + sub.sTime)
            )) AS current_group,
            CASE
                WHEN COALESCE(is_blocked, 0) = 1       THEN 0
                WHEN current_group LIKE '%LOST CARDS%' THEN 0
                ELSE 1
            END AS has_access,
            LEAD(timestamp) OVER (
                PARTITION BY lInstanceId ORDER BY timestamp, rn
            ) AS period_end
        FROM running_state
    ),
    
    -- ── 5. Gaps-and-islands collapse ──────────────────────────────────────────
    islands AS (
        SELECT
            timestamp,
            lInstanceId,
            CustomerName,
            current_group,
            is_blocked,
            has_access,
            period_end,
            ROW_NUMBER() OVER (PARTITION BY lInstanceId ORDER BY timestamp)
                - ROW_NUMBER() OVER (PARTITION BY lInstanceId, has_access ORDER BY timestamp)
                AS island_key
        FROM access_state
    )
    
    -- ── 6. Final aggregation ───────────────────────────────────────────────────
    SELECT
        csd.CardNumber PermitNumber,
        i.lInstanceId As CardID,
        MAX(i.CustomerName)   AS CustomerName,
        CASE
            WHEN MIN(i.timestamp) = MIN(MIN(i.timestamp)) OVER (PARTITION BY i.lInstanceId)
            THEN CAST(
                CASE
                    WHEN csd.effective_start_date > CAST(MIN(i.timestamp) AS DATE)
                    THEN csd.effective_start_date
                    ELSE CAST(MIN(i.timestamp) AS DATE)
                END
            AS DATETIME)
            ELSE MIN(i.timestamp)
        END                   AS period_start,
        MAX(i.period_end)     AS period_end,
        i.has_access,
        MIN(i.current_group)  AS group_at_start,
        MAX(i.is_blocked)     AS was_manually_blocked,
        csd.effective_start_date
    FROM islands i
    LEFT JOIN card_start_dates csd ON csd.lInstanceId = i.lInstanceId
    WHERE csd.CardNumber IS NOT NULL
    GROUP BY
        i.lInstanceId,
        csd.CardNumber,
        i.has_access,
        i.island_key,
        csd.effective_start_date
    ORDER BY
        csd.CardNumber,
        period_start;
    """, cnxn.get_engine('opms'))

permit_access_range['group_at_start'] = permit_access_range['group_at_start'].apply(lambda x: x.split(':')[1].lstrip())
permit_access_range['group_at_start'] = permit_access_range['group_at_start'].replace(group_name_changes)
permit_access_range.shape
```

```python
def merge_visits_with_policy(
    df: pd.DataFrame,
    permit_access_range: pd.DataFrame,
) -> pd.DataFrame:
    par = permit_access_range[permit_access_range["has_access"] == True][[
        "PermitNumber", "AccessPolicy", "GarageID",
        "period_start", "period_end"
    ]].copy()

    par["period_start"] = pd.to_datetime(par["period_start"])
    par["period_end"] = pd.to_datetime(par["period_end"])

    merged = df.merge(par, on="PermitNumber", how="left")

    mask = (
        (merged["EntryDate"] >= merged["period_start"]) &
        (merged["EntryDate"] <= merged["period_end"])
    )
    matched = merged[mask].copy()
    matched = matched.drop(columns=["period_start", "period_end"])
    matched = matched.drop_duplicates(subset=["PermitNumber", "EntryDate", "ExitDate"])

    # Unmatched: original visits whose PermitNumber+EntryDate+ExitDate
    # don't appear in matched at all
    match_key = matched.set_index(["PermitNumber", "EntryDate", "ExitDate"]).index
    orig_key = df.set_index(["PermitNumber", "EntryDate", "ExitDate"]).index
    unmatched = df[~orig_key.isin(match_key)]

    print(f"Visits matched to an active access period: {len(matched)}")
    print(f"Visits with no matching access period: {len(unmatched)}")

    return matched
```

**Target variable**  
For each (PermitNumber, Date) pair where the date falls within the permit's validity range and is a valid day per their access policy — did they show up (within their access window)? Binary 1/0.  
This means your training data isn't just the visits table — it's the full grid of permit-days, most of which are zeros. A permit active for 400 weekdays that showed up 120 times contributes 400 rows, not 120.

```python
# Suffixes that are billing-only and should be stripped before parsing
_BILLING_SUFFIXES = re.compile(
    r'\b(employee|pay|pass)\b', re.IGNORECASE
)

_DAY_MAP = {
    'M': 0, 'Tu': 1, 'W': 2, 'Th': 3, 'F': 4, 'Sa': 5, 'Su': 6
}

# All 7 days as a frozenset
_ALL_DAYS = frozenset(range(7))
_WEEKDAYS = frozenset(range(5))  # Mon–Fri


@dataclass(frozen=True)
class AccessPolicy:
    """
    Parsed representation of an access policy string.

    valid_days  : frozenset of weekday ints (0=Mon, 6=Sun) that count
                  toward the denominator. For M-Su / 24/7, this is all 7.
    window_start: int, minutes since midnight (inclusive)
    window_end  : int, minutes since midnight (exclusive)
    is_overnight: bool, True when window crosses midnight
                  (window_end < window_start)
    raw         : original policy string
    """
    valid_days: frozenset
    window_start: int   # minutes since midnight
    window_end: int     # minutes since midnight
    is_overnight: bool
    raw: str

    def is_valid_day(self, d: date) -> bool:
        return d.weekday() in self.valid_days

    def minute_in_window(self, minute: int) -> bool:
        """Is this minute (0–1439) within the access window?"""
        if self.is_overnight:
            return minute >= self.window_start or minute < self.window_end
        else:
            return self.window_start <= minute < self.window_end

    @property
    def window_description(self) -> str:
        s = f"{self.window_start // 60:02d}:{self.window_start % 60:02d}"
        e = f"{self.window_end // 60:02d}:{self.window_end % 60:02d}"
        return f"{s}–{e} {'(overnight)' if self.is_overnight else ''}"


def parse_access_policy(policy_str: str) -> AccessPolicy:
    """
    Parse an access policy string into an AccessPolicy dataclass.

    Handles patterns:
      [Garage] 24/7
      [Garage] M-F HH-HH
      [Garage] M-Su HH-HH
      [Garage] M-F HH-HH [billing suffix...]
    """
    # Strip billing suffixes and extra whitespace
    cleaned = _BILLING_SUFFIXES.sub('', policy_str).strip()

    # --- 24/7 policies ---
    if '24/7' in cleaned:
        return AccessPolicy(
            valid_days=_ALL_DAYS,
            window_start=0,
            window_end=1440,
            is_overnight=False,
            raw=policy_str,
        )

    # --- Day + hour window policies ---
    # Match: M-F 4-22  |  M-Su 17-8  |  M-F 18-6 & Sa-Su 0-23 (take first segment)
    # Strip anything after '&' — the '& Sa-Su 0-23' portion extends the window to
    # weekends; handle that as a second segment if needed.
    segments = [s.strip() for s in cleaned.split('&')]

    # Parse each segment into (valid_days, start_min, end_min)
    parsed_segments = []
    for seg in segments:
        m = re.search(r'(M-F|M-Su|Sa-Su)\s+(\d{1,2})-(\d{1,2})', seg)
        if not m:
            # Log and return a permissive fallback rather than crashing
            import logging
            logging.warning(f"Unrecognized policy format, defaulting to 24/7: '{policy_str}'")
            return AccessPolicy(
                valid_days=_ALL_DAYS,
                window_start=0,
                window_end=1440,
                is_overnight=False,
                raw=policy_str,
            )

        day_range, h_start, h_end = m.group(1), int(m.group(2)), int(m.group(3))

        if day_range == 'M-F':
            valid_days = _WEEKDAYS
        elif day_range == 'M-Su':
            valid_days = _ALL_DAYS
        elif day_range == 'Sa-Su':
            valid_days = frozenset({5, 6})
        else:
            raise ValueError(f"Unknown day range: {day_range}")

        start_min = h_start * 60
        end_min = h_end * 60
        is_overnight = end_min < start_min

        parsed_segments.append((valid_days, start_min, end_min, is_overnight))

    if len(parsed_segments) == 1:
        valid_days, start_min, end_min, is_overnight = parsed_segments[0]
    else:
        # Merge segments — union of valid days, assume same window hours
        # (e.g. 'M-F 18-6 & Sa-Su 0-23': treat as all-week overnight)
        all_valid = frozenset().union(*[s[0] for s in parsed_segments])
        # Use the first segment's window; the & segment typically extends days not hours
        _, start_min, end_min, is_overnight = parsed_segments[0]
        valid_days = all_valid

    return AccessPolicy(
        valid_days=valid_days,
        window_start=start_min,
        window_end=end_min,
        is_overnight=is_overnight,
        raw=policy_str,
    )
```

```python
def get_visit_dates(entry_dt: pd.Timestamp, exit_dt: pd.Timestamp) -> set[date]:
    """
    Return all dates spanned by a visit, inclusive of both entry and exit date.
    """
    entry_date = entry_dt.date()
    exit_date = exit_dt.date()
    n_days = (exit_date - entry_date).days + 1
    return {entry_date + timedelta(days=i) for i in range(n_days)}


def get_days_present(permit_df: pd.DataFrame) -> set[date]:
    """
    Return all dates where a permit was present, expanding multi-day visits.
    Only counts visits within the access agreement (Amount == 0).
    """
    valid_visits = permit_df[permit_df["Amount"] == 0]
    present = set()
    for _, row in valid_visits.iterrows():
        present |= get_visit_dates(row["EntryDate"], row["ExitDate"])
    return present

def compute_streak(prior_valid_days: list[date], visit_dates: set[date]) -> int:
    """
    Positive = number of consecutive most-recent valid days with a visit.
    Negative = number of consecutive most-recent valid days without a visit.
    0 = no prior valid days.
    """
    if not prior_valid_days:
        return 0

    # Walk backwards from most recent valid day
    last_status = prior_valid_days[-1] in visit_dates
    count = 0

    for past_day in reversed(prior_valid_days):
        if (past_day in visit_dates) == last_status:
            count += 1
        else:
            break

    return count if last_status else -count
```

```python
def build_permit_day_grid(
    visits: pd.DataFrame,
    permit_access_range: pd.DataFrame,
    date_start: date,
    date_end: date,
) -> pd.DataFrame:
    """
    Build the (PermitNumber, Date) grid for all permits using
    permit_access_range to determine valid date ranges.
    
    Only days where has_access == True are included in the grid.
    Each row is one valid permit-day, labeled 1 if the permit holder
    showed up within their access window, 0 otherwise.
    """
    records = []

    # Only visits within access agreement
    valid_visits = visits.copy()
    valid_visits["Date"] = valid_visits["EntryDate"].dt.date

    for permit_num, access_df in permit_access_range.groupby("PermitNumber"):
        # Only active periods
        active_periods = access_df[access_df["has_access"] == 1]
        if active_periods.empty:
            continue

        policy_str = active_periods["AccessPolicy"].mode()[0]
        policy = parse_access_policy(policy_str)

        group = active_periods["GroupName"].mode()[0]
        permit_name = active_periods["CustomerName"].mode()[0]

        # Build visit_dates from actual visits for this permit
        permit_visits = valid_visits[valid_visits["PermitNumber"] == permit_num]
        visit_dates = set()
        for _, row in permit_visits.iterrows():
            visit_dates |= get_visit_dates(row["EntryDate"], row["ExitDate"])
        visit_dates = {d for d in visit_dates if d.weekday() in policy.valid_days}

        # Walk each active access period, clamped to the requested date range
        for _, period in active_periods.iterrows():
            period_start = max(period["period_start"].date(), date_start)
            period_end = min(period["period_end"].date(), date_end)

            if period_start > period_end:
                continue

            n_days = (period_end - period_start).days + 1
            for i in range(n_days):
                d = period_start + timedelta(days=i)
                if d.weekday() not in policy.valid_days:
                    continue

                records.append({
                    "PermitNumber": permit_num,
                    "PermitName": permit_name,
                    "GroupName": group,
                    "AccessPolicy": policy_str,
                    "Date": d,
                    "DayOfWeek": d.weekday(),
                    "Month": d.month,
                    "WeekOfYear": d.isocalendar().week,
                    "IsHoliday": is_holiday(d),
                    "showed_up": int(d in visit_dates),
                })

    return pd.DataFrame(records)


HOLIDAYS = {
    # Add your municipal holidays here as (month, day) tuples
    # These apply every year
    (1, 1),   # New Year's Day
    (6, 19),  # Juneteenth
    (11, 11), # Veterans Day
    (12, 24), # Christmas Eve
    (12, 25), # Christmas Day
    (12, 31), # New Year's Eve
}

FLOATING_HOLIDAYS = {
    # MLK Day: 3rd Monday of January
    #(1,19),
    # Memorial Day: last Monday of May
    # Labor Day: 1st Monday of September
    # Thanksgiving: 4th Thursday of November
    # Day after Thanksgiving: 4th Thursday + 1
    # These need date-specific logic
}

def is_holiday(d: date) -> bool:
    if (d.month, d.day) in HOLIDAYS:
        return True
    # Thanksgiving: 4th Thursday of November
    if d.month == 11 and d.weekday() == 3:
        if 22 <= d.day <= 28:
            return True
    # Day after Thanksgiving
    if d.month == 11 and d.weekday() == 4:
        if 23 <= d.day <= 29:
            return True
    # MLK Day: 3rd Monday of January
    if d.month == 1 and d.weekday() == 0:
        if 15 <= d.day <= 21:
            return True
    # Memorial Day: last Monday of May
    if d.month == 5 and d.weekday() == 0:
        if d.day >= 25:
            return True
    # Labor Day: 1st Monday of September
    if d.month == 9 and d.weekday() == 0:
        if d.day <= 7:
            return True
    return False
```

```python
def add_historical_features(
    grid: pd.DataFrame,
    visits: pd.DataFrame,
) -> pd.DataFrame:
    """
    Add per-permit historical features to the grid.
    All lookback windows are computed from data strictly before each row's Date.
    """
    grid = grid.sort_values(["PermitNumber", "Date"]).copy()

    # Valid visits only (within access agreement)
    valid_visits = visits.copy()
    valid_visits["Date"] = valid_visits["EntryDate"].dt.date

    feature_rows = []

    for permit_num, permit_grid in grid.groupby("PermitNumber"):
        permit_visits = valid_visits[valid_visits["PermitNumber"] == permit_num]
        
        policy_str = permit_grid["AccessPolicy"].iloc[0]
        policy = parse_access_policy(policy_str)

        visit_dates = set()
        for _, row in permit_visits.iterrows():
            visit_dates |= get_visit_dates(row["EntryDate"], row["ExitDate"])
        visit_dates = {d for d in visit_dates if d.weekday() in policy.valid_days}
        
        for _, row in permit_grid.iterrows():
            d = row["Date"]

            # History strictly before this date
            prior_dates = sorted([v for v in visit_dates if v < d])
            prior_valid_days = [
                r for r in permit_grid["Date"] if r < d
            ]

            n_prior_valid = len(prior_valid_days)
            n_prior_visits = len(prior_dates)

            # Overall presence rate
            presence_rate = n_prior_visits / n_prior_valid if n_prior_valid > 0 else np.nan

            # Trailing window presence rates
            def visits_in_last_n_valid_days(n):
                last_n = prior_valid_days[-n:] if len(prior_valid_days) >= n else prior_valid_days
                if not last_n:
                    return np.nan
                hits = sum(1 for d_ in last_n if d_ in visit_dates)
                return hits / len(last_n)

            # Day-of-week presence rate
            dow = d.weekday()
            dow_valid = [r for r in prior_valid_days if r.weekday() == dow]
            dow_visits = [r for r in prior_dates if r.weekday() == dow]
            dow_rate = len(dow_visits) / len(dow_valid) if dow_valid else np.nan

            # Compute streak
            streak = compute_streak(prior_valid_days, visit_dates)

            # Days since last visit
            days_since = (d - prior_dates[-1]).days if prior_dates else np.nan

            # Same day last week
            last_week = d - timedelta(days=7)
            visited_last_week = int(last_week in visit_dates)

            feature_rows.append({
                **row.to_dict(),
                "PresenceRate": presence_rate,
                "PresenceRateLast10": visits_in_last_n_valid_days(10),
                "PresenceRateLast20": visits_in_last_n_valid_days(20),
                "DayOfWeekRate": dow_rate,
                "Streak": streak,
                "DaysSinceLastVisit": days_since,
                "VisitedSameLastWeek": visited_last_week,
                "PermitTenureDays": (d - permit_grid["Date"].min()).days,
            })

    return pd.DataFrame(feature_rows)
```

```python
FEATURES = [
    "DayOfWeek", "Month", "WeekOfYear", "IsHoliday",
    "PresenceRate", "PresenceRateLast10", "PresenceRateLast20",
    "DayOfWeekRate", "DaysSinceLastVisit",
    "VisitedSameLastWeek", #"PermitTenureDays", "Streak",
]

# Temporal split boundaries.
# Train on the first ~6 months; validate on the next ~3 months (used for
# calibration only); hold the remaining window as a true future test set.
TRAIN_CUTOFF = date(2025, 9, 30)   # last day of training window
VAL_CUTOFF   = date(2025, 12, 31)  # last day of calibration window

def train_model(
    features: pd.DataFrame,
    train_cutoff: date = TRAIN_CUTOFF,
    val_cutoff: date = VAL_CUTOFF,
):
    df = features.dropna(subset=FEATURES).copy()

    train_df = df[df["Date"] <= train_cutoff]
    val_df   = df[(df["Date"] > train_cutoff) & (df["Date"] <= val_cutoff)]
    test_df  = df[df["Date"] > val_cutoff]

    X_train, y_train = train_df[FEATURES], train_df["showed_up"]
    X_val,   y_val   = val_df[FEATURES],   val_df["showed_up"]
    X_test,  y_test  = test_df[FEATURES],  test_df["showed_up"]

    print(f"Train : {len(train_df):>6,} rows  {train_df['Date'].min()} → {train_df['Date'].max()}")
    print(f"Val   : {len(val_df):>6,} rows  {val_df['Date'].min()} → {val_df['Date'].max()}")
    print(f"Test  : {len(test_df):>6,} rows  {test_df['Date'].min()} → {test_df['Date'].max()}")
    print()

    model = lgb.LGBMClassifier(
        n_estimators=500,
        learning_rate=0.05,
        num_leaves=31,
        #class_weight="balanced",
        random_state=42,
    )
    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    print(classification_report(y_test, y_pred))
    print(f"ROC-AUC (future test window): {roc_auc_score(y_test, y_prob):.3f}")

    return model, X_train, X_val, X_test, y_train, y_val, y_test
```

**Temporal train/test split** — the model is trained on Apr–Sep 2025, then evaluated on Jan–present 2026. This mirrors the actual prediction task: given past behavior, predict future show-ups. The earlier group-shuffle split evaluated on the same time window the model trained on, which inflated scores.

**Validation window (Oct–Dec 2025)** — held out from training and used only to fit calibration. Not used for final metric reporting.

**Cold start** — permits with fewer than ~10 valid days of history will have unreliable habit features. Consider flagging these and falling back to group-level averages.

**Class imbalance** — presence rates near 50% are roughly balanced. If your presence rates drop (seasonal low-use periods), consider enabling `class_weight="balanced"` or tuning the classification threshold away from 0.5.


```python
model, X_train, X_val, X_test, y_train, y_val, y_test = train_model(features)
```

```python
def build_prediction_row(
    permit_num: str,
    permit_history: pd.DataFrame,  # all historical visits for this permit
    prediction_date: date,
    policy: AccessPolicy,
    period_start: date,
) -> dict:
    """
    Build a single feature row for a given permit and prediction date.
    Uses only history strictly before prediction_date.
    """
    valid_visits = permit_history.copy()
    
    # Expand multi-day visits and filter to valid days
    visit_dates = set()
    for _, row in valid_visits.iterrows():
        visit_dates |= get_visit_dates(row["EntryDate"], row["ExitDate"])
    visit_dates = {d for d in visit_dates if d.weekday() in policy.valid_days}

    # All valid days before the prediction date
    first_seen = period_start
    prior_valid_days = [
        first_seen + timedelta(days=i)
        for i in range((prediction_date - first_seen).days)
        if (first_seen + timedelta(days=i)).weekday() in policy.valid_days
    ]

    n_prior_valid = len(prior_valid_days)
    n_prior_visits = len([d for d in visit_dates if d < prediction_date])

    presence_rate = n_prior_visits / n_prior_valid if n_prior_valid > 0 else np.nan

    def visits_in_last_n_valid_days(n):
        last_n = prior_valid_days[-n:] if len(prior_valid_days) >= n else prior_valid_days
        if not last_n:
            return np.nan
        hits = sum(1 for d in last_n if d in visit_dates)
        return hits / len(last_n)

    dow = prediction_date.weekday()
    dow_valid = [d for d in prior_valid_days if d.weekday() == dow]
    dow_visits = [d for d in visit_dates if d < prediction_date and d.weekday() == dow]
    dow_rate = len(dow_visits) / len(dow_valid) if dow_valid else np.nan

    last_week = prediction_date - timedelta(days=7)
    days_since = (
        prediction_date - max((d for d in visit_dates if d < prediction_date), default=None)
    ).days if any(d < prediction_date for d in visit_dates) else np.nan

    return {
        "PermitNumber": permit_num,
        "DayOfWeek": dow,
        "Month": prediction_date.month,
        "WeekOfYear": prediction_date.isocalendar().week,
        "IsHoliday": is_holiday(prediction_date),
        "PresenceRate": presence_rate,
        "PresenceRateLast10": visits_in_last_n_valid_days(10),
        "PresenceRateLast20": visits_in_last_n_valid_days(20),
        "DayOfWeekRate": dow_rate,
        "Streak": compute_streak(prior_valid_days, visit_dates),
        "DaysSinceLastVisit": days_since,
        "VisitedSameLastWeek": int(last_week in visit_dates),
        "PermitTenureDays": (prediction_date - first_seen).days,
    }


def predict_tomorrow(
    active_permits: pd.DataFrame,  # filtered to currently active permits
    all_visits: pd.DataFrame,
    model: lgb.LGBMClassifier,
    prediction_date: date = None,
) -> pd.DataFrame:
    if prediction_date is None:
        prediction_date = date.today() + timedelta(days=1)

    active_permit_numbers = set(active_permits["PermitNumber"])
    relevant_visits = all_visits[all_visits["PermitNumber"].isin(active_permit_numbers)]

    rows = []
    for permit_num, permit_history in relevant_visits.groupby("PermitNumber"):
        policy_str = permit_history["AccessPolicy"].mode()[0]
        policy = parse_access_policy(policy_str)
    
        if prediction_date.weekday() not in policy.valid_days:
            continue
    
        # Get the earliest active period start for this permit
        permit_access = permit_access_range[
            (permit_access_range["PermitNumber"] == permit_num) &
            (permit_access_range["has_access"] == True)
        ]
        if permit_access.empty:
            continue
        
        period_start = permit_access["period_start"].min().date()
    
        row = build_prediction_row(
            permit_num, permit_history, prediction_date, policy, period_start
        )
        rows.append(row)

    if not rows:
        return pd.DataFrame()

    pred_df = pd.DataFrame(rows)
    pred_df["DaysSinceLastVisit"] = pred_df["DaysSinceLastVisit"].astype(float)
    X = pred_df[FEATURES]

    print(pred_df[FEATURES].isna().sum())
    print(pred_df[FEATURES].describe())
    pred_df["ShowUpProbability"] = model.predict_proba(X)[:, 1]
    pred_df["PredictedShowUp"] = (pred_df["ShowUpProbability"] >= 0.5).astype(int)

    return pred_df.sort_values(
        "ShowUpProbability", ascending=False
    )
```

```python
from sklearn.calibration import calibration_curve
import matplotlib.pyplot as plt

# Evaluate calibration on the held-out test window (Jan–present).
# A well-calibrated model should lie close to the diagonal.
y_test_prob = model.predict_proba(X_test)[:, 1]
fraction_of_positives, mean_predicted = calibration_curve(y_test, y_test_prob, n_bins=10)

plt.figure(figsize=(6, 5))
plt.plot(mean_predicted, fraction_of_positives, marker='o', label='Model (raw)')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect calibration')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration curve — future test window')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.calibration import CalibratedClassifierCV, calibration_curve
import matplotlib.pyplot as plt

# Fit isotonic calibration on the validation window (Oct–Dec 2025).
# This window is separate from the test set so calibration is not circular.
calibrated = CalibratedClassifierCV(model, method='isotonic', cv='prefit')
calibrated.fit(X_val, y_val)

# Compare raw vs calibrated probabilities on the held-out test window.
y_test_prob_raw = model.predict_proba(X_test)[:, 1]
y_test_prob_cal = calibrated.predict_proba(X_test)[:, 1]

frac_raw, pred_raw = calibration_curve(y_test, y_test_prob_raw, n_bins=10)
frac_cal, pred_cal = calibration_curve(y_test, y_test_prob_cal, n_bins=10)

plt.figure(figsize=(6, 5))
plt.plot(pred_raw, frac_raw, marker='o', label='Raw')
plt.plot(pred_cal, frac_cal, marker='s', label='Isotonic calibrated')
plt.plot([0, 1], [0, 1], linestyle='--', label='Perfect')
plt.xlabel('Mean predicted probability')
plt.ylabel('Fraction of positives')
plt.title('Calibration: raw vs isotonic — future test window')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Raw prob sum on test set:        {y_test_prob_raw.sum():.1f}")
print(f"Calibrated prob sum on test set: {y_test_prob_cal.sum():.1f}")
print(f"Actual positives in test set:    {y_test.sum()}")
```
